```python
import pandas as pd
```

```python
data = pd.read_excel('prediction.xlsx','titanic')
data.head()
```

```python
import numpy as np
```

```python
data1= data['Age'].fillna(data['Age'].mean())
```

```python
data1.head(20)
```

```python
data['Age']= data1
```

```python
data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex =LabelEncoder()
```

```python
data['Sex_n']= le_sex.fit_transform(data.Sex)
data.head()
```

```python
data = data.drop(['Sex'],axis = 1)
data.head()
```

```python
x = data.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 1)
```

```python
len(x)
```

```python
y = data.Survived
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .2)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
